```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread('/content/gdrive/MyDrive/messi.jpg')
```

```
plt.imshow(img), plt.title('orijinal')
```

```
genislik,uzunluk = img.shape[:2]
```

```
M = np.float32([[1,0,50],[0,1,50]])
kaydirma = cv2.warpAffine(img,M,(uzunluk,genislik))
```

```
plt.subplot(122), plt.imshow(kaydirma), plt.title('Kaydırılmış')
plt.show()
```

```
satir, sutun = img.shape[:2]
```

```
M = cv2.getRotationMatrix2D((sutun/2,satir/2),45,1)
cevirme = cv2.warpAffine(img,M,(sutun,satir))
```

```
plt.subplot(121), plt.imshow(img), plt.title('orjinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(cevirme), plt.title('çevrilmiş')
plt.xticks([]), plt.yticks([])

plt.show()
```

```
img = cv2.imread('/content/gdrive/MyDrive/affine.jpg')
```

```
satir, sutun = img.shape[:2]
```

```
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
```

```
M = cv2.getAffineTransform(pts1,pts2)
```

```
hassas = cv2.warpAffine(img, M, (sutun,satir))
```

```
plt.subplot(121), plt.imshow(img), plt.title('orjinal')
plt.subplot(122), plt.imshow(hassas), plt.title('Hassas donusum')
plt.show()
```

```
img = cv2.imread('/content/gdrive/MyDrive/perspective.jpg')
```

```
plt.subplot(121),plt.imshow(img), plt.title('orijinal')
plt.show()
```

```
pts1 = np.float32([[25,25],[185,25],[10,195],[195,195]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
```

```
M = cv2.getPerspectiveTransform(pts1,pts2)
```

```
pers = cv2.warpPerspective(img,M,(200,200))
```

```
plt.subplot(121), plt.imshow(img), plt.title('orijinal')
plt.subplot(122), plt.imshow(pers), plt.title('Perspektif Düzeltme')
plt.show()
```
