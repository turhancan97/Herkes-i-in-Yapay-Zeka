```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv.imread('suduko.jpg',0)
kenar = cv.Canny(img,50,90)
```

```
plt.figure(figsize=(12,14))

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(kenar, cmap='gray'), plt.title('Kenarlar')
plt.xticks([]), plt.yticks([])

plt.show()
```

```
yuz_img = cv.imread('kız cocugu.png',0)
yuz_kenar_img = cv.Canny(yuz_img, 50, 100)
```

```
plt.figure(figsize=(12,16))

plt.subplot(121), plt.imshow(yuz_img, cmap='gray'), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(yuz_kenar_img, cmap='gray'), plt.title('Kenar')
plt.xticks([]), plt.yticks([])

plt.show()
```
