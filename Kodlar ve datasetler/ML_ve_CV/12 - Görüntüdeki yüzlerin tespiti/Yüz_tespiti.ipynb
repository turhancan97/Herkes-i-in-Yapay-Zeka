```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
resim = cv2.imread('/content/gdrive/MyDrive/yuz.jpeg')
```

```
plt.imshow(resim)
plt.xticks([]), plt.yticks([])
plt.show()
```

```
yuz_cascade = cv2.CascadeClassifier("/content/gdrive/MyDrive/haarcascade_frontalface_default.xml")
```

```
goz_cascade = cv2.CascadeClassifier("/content/gdrive/MyDrive/haarcascade_eye.xml")
```

```
yuzler = yuz_cascade.detectMultiScale(resim, scaleFactor=1.075, minNeighbors=5, minSize=(15,15))
```

```
for(x,y,w,h) in yuzler:
  cv2.rectangle(resim, (x,y),(x+w,y+h), (255,0,0),2 )
  roi = resim[y:y+h, x:x+w]
  gozler = goz_cascade.detectMultiScale(roi)
  for(ex,ey,ew,eh) in gozler:
    cv2.rectangle(roi, (ex,ey), (ex+ew, ey+eh), (0,255,0),1)
```
