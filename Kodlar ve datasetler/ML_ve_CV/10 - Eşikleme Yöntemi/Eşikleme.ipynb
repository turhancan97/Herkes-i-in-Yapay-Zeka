```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
kitap_img = cv.imread('kitapsayfasi.jpg')
kitap_img = cv.cvtColor(kitap_img, cv.COLOR_BGR2RGB)
```

```
parlak_img = cv.convertScaleAbs(kitap_img,beta=100)
```

```
plt.figure(figsize=(15,20))
plt.subplot(121), plt.imshow(kitap_img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(parlak_img), plt.title('Parlak')
plt.xticks([]), plt.yticks([])

plt.show()
```

```
kitap_img = cv.cvtColor(kitap_img, cv.COLOR_RGB2GRAY)
```

```
ret, thresh_img = cv.threshold(kitap_img,5,255, cv.THRESH_BINARY)

plt.figure(figsize=(12,8))
plt.imshow(thresh_img,cmap='gray')
```

```
adaptive_thresh_img = cv.adaptiveThreshold(kitap_img,255,
                                           cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,15,2)

plt.figure(figsize=(12,8))
plt.imshow(adaptive_thresh_img, cmap='gray')
```
